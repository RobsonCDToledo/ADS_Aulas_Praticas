# Visualização de dados em Python

Aplicando as bibliotecas:  
* matplotlib
* pandas
* seaborn



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
# Exemplo simples utilizando matplotlib

# Gerando lista de dados aleatórios

dados1 = random.sample(range(100), k=20)
dados2 = random.sample(range(100), k=20)

plt.plot(dados1, dados2)
```

```python
# Exemplo utilizando função plot() do pandas

dados = {
    'nome' : ['Produto A', 'Produto B', 'Produto C', 'Produto A', 'Produto E'],
    'qtd_comprada':[3, 1, 4, 3 ,2],
    'tipo_de_item': ['Eletrônico', 'Vestuário', 'Alimento', 'Eletrônico', 'Alimento'],
    'receita_total': [120, 80, 60, 120, 90]
    }

df = pd.DataFrame(dados)
df.drop_duplicates(keep='last', inplace=True)
df.plot(x='nome', y='receita_total', kind='barh', title='Vendas por Produto', )
```

```python
df.plot(x='nome', y='receita_total', kind='pie', title='Vendas por Produto', )
```

```python
df.plot(x='nome', y='receita_total', kind='bar', title='Vendas por Produto', )
```

```python
df.plot(x='nome', y='receita_total', kind='line', title='Vendas por Produto', )
```

```python
df.plot(x='nome', y='receita_total', kind='scatter', title='Vendas por Produto', )
```

# Exemplo de Gráficos utilizando o SEABORN

```python
sns.set(style='whitegrid')
df_tips = sns.load_dataset('tips')
```

```python
# Gerando um painel de gráficos 

fig, ax = plt.subplots(1, 3, figsize=(15,5))

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[0])

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[1], estimator=sum)

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[2], estimator=len)
```
